```python
#Dependancies
import glob
import pandas as pd
import os
```

```python
# Folder path
path = 'Resources'
all_files = glob.glob(path + "/*.csv")

# DF where all csv files will be loaded
data = []

# Loop to load all CSV file dataframes
for csv in all_files:
    frame = pd.read_csv(csv, index_col=None, header=0)
    # Add new column with the 
    frame['filename'] = os.path.basename(csv)
    data.append(frame)
```

```python
# Pull column names from the collected data frames into a new blank DF
# Done to facilitate later merge of multiple DFs
cols = data[1].columns.values
bike_df = pd.DataFrame(columns = cols)
bike_df
```

```python
# Merge all Files
for x in range(0, len(data)):
    bike_df = bike_df.merge(data[x], on=None, how='outer')

bike_df
```

```python
# Verified all files loaded into the single DF
bike_df['filename'].unique()
```

```python
#Create date column from filename column
bike_df['report month'] = bike_df.apply(lambda row: f"{row.filename[7:9]}/{row.filename[3:7]}" , axis = 1) 
bike_df
```

```python
# Identify if there is missing data
bike_df.count()
```

```python
# Evaluate data types
bike_df.dtypes
```

```python
# Converting gender numeric values to text values
bike_df['gender']= bike_df['gender'].astype(str)
bike_df.dtypes
```

```python
# # Created empty gender column for text values
bike_df['gender2'] = ''
bike_df

#Change Gender values
#Reference from site: https://www.citibikenyc.com/system-data
#Gender (Zero=unknown; 1=male; 2=female)
for x in range(len(bike_df.gender)):
    if bike_df.gender[x] =='1':
        bike_df['gender2'][x] ='male'
    elif bike_df.gender[x] == '2':
        bike_df['gender2'][x] ='female'
    else:
        bike_df['gender2'][x] = 'unknown'
     
bike_df
```

```python
# Drop unneeded column and rename new column
bike_df = bike_df.drop(columns=['gender'], axis = 1)
bike_df = bike_df.rename(columns={'gender2': 'gender'})
bike_df
```

```python
# Create date column that is extracted from starttime column
bike_df['date'] = bike_df['starttime'][0:10]
```

```python
# Change startime and stoptime columns to date time format
bike_df['starttime'] = pd.to_datetime(bike_df['starttime'], format='%Y-%m-%d %H:%M:%S')
bike_df['stoptime'] = pd.to_datetime(bike_df['starttime'], format='%Y-%m-%d %H:%M:%S')
bike_df.dtypes
```

```python
# Replaced date column with date information from start time column to keep date format
bike_df['date'] = bike_df['starttime'].dt.date
```

```python
bike_df
```

```python
# Trip duration is in seconds
bike_df['tripduration'] = bike_df['tripduration']/60
bike_df
```

```python
# Dropped unneeded columns
bike_df = bike_df.drop(['filename', 'report month'], axis=1)
```

```python
bike_df
```

```python
# Updated Lat and long values to facilitate table merging
bike_df[['start station latitude', 
        'start station longitude', 
        'end station latitude', 
        'end station longitude']] = (bike_df[['start station latitude', 
                                             'start station longitude', 
                                             'end station latitude', 
                                             'end station longitude']].round(decimals=5))
#     ataFrame.round(decimals=0, *args, **kwargs)

# df[["String 1", "String 2"]] = df[["String 1", "String 2"]].apply(prepend_geek) 
bike_df
```
