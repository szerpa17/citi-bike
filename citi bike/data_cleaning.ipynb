```python
#Dependancies
import glob
import pandas as pd
import os
```

```python
# Folder path
path = 'Resources'
all_files = glob.glob(path + "/*.csv")

# DF where all csv files will be loaded
data = []

# Loop to load all CSV file dataframes
for csv in all_files:
    frame = pd.read_csv(csv, index_col=None, header=0)
    # Add new column with the 
    frame['filename'] = os.path.basename(csv)
    data.append(frame)
```

```python
# Pull column names from the collected data frames into a new blank DF
# Done to facilitate later merge of multiple DFs
cols = data[1].columns.values
bike_df = pd.DataFrame(columns = cols)
bike_df
```

```python
# Merge all Files
for x in range(0, len(data)):
    bike_df = bike_df.merge(data[x], on=None, how='outer')

bike_df
```

```python
# Verified all files loaded into the single DF
bike_df['filename'].unique()
```

```python
#Create date column from filename column
bike_df['report month'] = bike_df.apply(lambda row: f"{row.filename[7:9]}/{row.filename[3:7]}" , axis = 1) 
bike_df
```
